```python

import pandas as pd 
#2
import seaborn as sns
```

```python
#1
#Tạo hàm load data với tham số titanic209
#Dùng read_csv để đọc dữ liệu từ file csv và truyền vào biến titanic209
def load_data(titanic209):
  return pd.read_csv(titanic209)
```

```python
#Gán biến dữ liệu để chứa file
titanic209 = load_data("titanic_disaster.csv")
```

```python
/# in 10 dòng đầu
titanic209.head(10)
```

```python
#2
#sns.heatmap vễ biểu đồ nhiệt headmap
#titanic209['Age'].isna() trả về một giá trị boolean = True ở hàng mà cột Age thiếu
#titanic209[titanic209['Age'].isna()] lọc và trả về các hàng trong titanic209 mà cột Age thiếu
#print(dfage.count()) -> đếm số lượng giá trị không bị thiếu  tr ong mỗi cộtdfage.

sns.heatmap(titanic209.isna())
dfage=titanic209[titanic209['Age'].isna()]
print(dfage.count())
dfcabin=titanic209[titanic209['Cabin'].isna()]
print(dfcabin.count())
dfembarked=titanic209[titanic209['Embarked'].isna()]
print(dfembarked.count())
```

```python
#3
#Tách cột name thành 2 cột 
titanic209[['Firstname', 'secondName']]=titanic209['Name'].str.split(",",expand = True)
```

```python
#in
titanic209
```

```python
#Xóa cột name
titanic209 = titanic209.drop('Name',axis=1)
titanic209
```

```python
#4
#Rút gọn size trên cột sex ( Male -> M & Female -> F)
#Sử dụng replace để thay thế các values

titanic209['Sex'] = titanic209['Sex'].replace({'male': 'M', 'female': 'F'})
titanic209
```

```python
#5
# Vẽ biểu đồ Box plot cho biến Age theo từng nhóm Pclass
sns.boxplot(y='Age', x='Pclass', data =titanic209)
```

```python
#5a
#mean tính gt trung bình
#round thì làm tròn giá trị
#fillna là điền giá trị thiếu trong cột 
titanic209['Age'].fillna(round(titanic209['Age'].mean()), inplace= True)
titanic209.head(10)
```

```python
#5b
#Groupby nhóm các hàng theo cợt pclass
#['Age'].median() tính giá trị trung vị 

median_ages = titanic209.groupby('Pclass')['Age'].median()
```

```python
#apply hàm tùy chỉnh cho từng hàng 
#lambda x: median_ages[x['Pclass']] if pd.isnull(x['Age']) else x['Age'] hàm tùy chỉnh 'lambda để kiểm tra từng hàng ' x' 
#if pd.isnull(x['Age']) else x['Age'], axis=1) - isnull kiểm tra nếu gt Age bị thiếu  - else ... Nếu gtri kh bị thiếu thì giữ nguyên gtri đó 

titanic209['Age'] = titanic209.apply(lambda x: median_ages[x['Pclass']] if pd.isnull(x['Age']) else x['Age'], axis=1)
titanic209 
```

```python
# Tạo biểu đồ nhiệt để nhận biết và kiểm tra dự liệu
sns.heatmap(titanic209.isna()) 
            
```

```python
#6
#sử dụng hàm loc để trích xuất dlieu và gán dk 
titanic209.loc[(titanic209['Age'] <= 12), 'Agegroup'] = 'Kid'
titanic209.loc[(titanic209['Age'] > 12) & (titanic209['Age'] <= 18), 'Agegroup'] = 'Teen'
titanic209.loc[(titanic209['Age'] > 18) & (titanic209['Age'] <= 60), 'Agegroup'] = 'Adult'
titanic209.loc[(titanic209['Age'] > 60), 'Agegroup'] = 'Older'
titanic209.head(10)
```

```python
#7
#extract(r'(Mr\.|Mrs\.|Miss\.|Master\.)', expand=False): Áp dụng hàm extract để trích xuất chuỗi phù hợp với biểu thức chính quy r'(Mr\.|Mrs\.|Miss\.|Master\.)
titanic209['namePrefix'] = titanic209['secondName'].str.extract(r'(Mr\.|Mrs\.|Miss\.|Master\.)', expand=False)
print(titanic209[['secondName', 'namePrefix']].head(10))
```

```python
#in
titanic209
```

```python
#8 
# Khai thác thêm thông tin số lượng thành viên đi theo nhóm thân quen (familySize) đối
#với mỗi hành khách trên chuyến hải trình; family size = 1+ SibSp + Parch
# Gán Sibsp = anh chị em, vck trên tàu 
# Gán Parch = cha mẹ, con cái trên tàu
titanic209['familySize'] = 1 + titanic209['SibSp'] + titanic209['Parch']
titanic209.head()
```

```python
#9
# Dùng vòng lặp for để thỏa dk 
# Tạo thêm đặc trưng ‗Alone‘ để xác định hành khách đi theo nhóm hay cá nhân bằng cách
#dựa trên familySize như sau: Nếu familySize = 0 thì giá trị Alone = 1 và ngược lại là 0.
for i in titanic209['familySize']:
  if i == 0:
    titanic209['Alone'] = 1
  else:
    titanic209['Alone'] = 0
```

```python
#10
#Tiến hành tách loại cabin (typeCabin) mà hành khách ở để lọc và phân tích đặc tính cabin
#Loại cabin được kí hiệu bởi chữ cái đầu tiên. Lưu ý: Đối với dữ liệu cabin bị thiếu thì thay thế bằng ―Unknown‖

#Tạo cột mới typeCabin dc lấy gtri từ cột Cabin , nếu Cabin 0 có gtri = Unknown
#For i in range được sử dụng để duyệt từng hàng của cột
#str(typeCabin[i]) chuyển thành giá trị chuỗi -> temp
#if temp != 'Unknown': typeCabin[i] = temp[:1]: Điều kiện kiểm tra nếu temp #'Unknown', tức là cột Cabin không phải là NaN, thì giá trị của typeCabin[i] sẽ được
# cắt lấy chỉ ký tự đầu tiên của temp và gán lại vào typeCabin[i].
titanic209['typeCabin'] = titanic209['Cabin'].fillna('Unknown')
typeCabin = titanic209['typeCabin']
for i in range(len(typeCabin)):
  temp = str(typeCabin[i])
  if temp != 'Unknown':
    typeCabin[i] = temp[:1]

```
