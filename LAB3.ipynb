```python
#problem 1
#them header va dataframe de dien giai du lieu 
column_names=["Id", "Name", "Age","Weight",'m0006', 'm0612', 'm1218','f0006','f0612','f1218']
#Doc file du lieu
patientheartrate = pd.read_csv("patient_heart_rate.csv", names = column_names)
```

```python
#hien thi vai dong du lieu dau tien ra man hinh 
print(patientheartrate.head(5))
```

```python
#tách cột 'Name' thành hai cột mới 'Firstname' và 'Lastname' trong DataFrame patientheartrate.
#Problem 2
patientheartrate[['Firstname', 'Lastname']]=patientheartrate['Name'].str.split(expand = True)
patientheartrate
```

```python
#xóa cột name
patientheartrate = patientheartrate.drop('Name',axis=1)
patientheartrate
```

```python

#Problem 3
#chuyển đổi từ pounds sang kilograms trong cột 'Weight' của patientheartrate.
weight=patientheartrate['Weight']
for i in range(len(weight)):
  x = str(weight[i])
  if "lbs" in x[-3:]:
   
    x = x[:-3]
    float_x = float(x)
    y = int(float_x/2.205)
    y = str(y)+"kgs"
    weight[i] = y
print(patientheartrate.head(5))
```

```python
#Problem 4
# loại bỏ các hàng chỉ chứa giá trị NaN
patientheartrate.dropna(how="all", inplace=True)
patientheartrate
```

```python
#Problem 5
#xử lý và loại bỏ các hàng trùng lặp dựa trên các cột được chỉ định
patientheartrate = patientheartrate.drop_duplicates(subset=['Firstname', 'Lastname','Age','Weight'])
print(patientheartrate)
```

```python
#Problem6
#làm sạch các ký tự không phải ASCII trong các cột 'Firstname' và 'Lastname'
#giúp chuẩn hóa dữ liệu và loại bỏ các ký tự không mong muốn.
patientheartrate.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
patientheartrate.Lastname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print(patientheartrate)
```

```python
#Problem 7
# điền các giá trị thiếu trong cột 'Age' bằng giá trị trung bình của cột đó và lưu trực tiếp vào DataFrame.
patientheartrate['Age'].fillna(patientheartrate['Age'].mean(), inplace =False)
```

```python
#biến đổi patientheartrate thành dạng dài (long format) và sắp xếp các hàng theo thứ tự của các cột 'Id', 'Age', 'Weight', 'Firstname', 'Lastname'.
patientheartrate_ = pd.melt(patientheartrate, id_vars=['Id', 'Age', 'Weight', 'Firstname', 'Lastname'], value_name="PulseRate",var_name="sex_and_time").sort_values(['Id', 'Age', 'Weight', 'Firstname', 'Lastname'])
patientheartrate_
```

```python
# lưu vào DataFrame tmp_patientheartrate, mỗi nhóm trong biểu thức chính quy sẽ là một cột trong DataFrame mới.
tmp_patientheartrate= patientheartrate_["sex_and_time"].str.extract("(\D)(\d+)(\d{2})", expand=True)
```

```python
tmp_patientheartrate
```

```python
# DataFrame tmp_patientheartrate đã được đổi tên các cột thành "Sex", "hours_lower" và "hours_upper".
tmp_patientheartrate.columns=["Sex","hours_lower","hours_upper"]
tmp_patientheartrate
```

```python
#thêm cột "Time" vào tmp_patientheartrate chứa giá trị từ các cột "hours_lower" và "hours_upper" được nối với dấu gạch ngang (-) giữa chúng.
tmp_patientheartrate["Time"]=tmp_patientheartrate["hours_lower"] + "-"+ tmp_patientheartrate["hours_upper"]
tmp_patientheartrate
```

```python
#tạo ra một DataFrame mới patientheartrate.
patientheartrate =pd.concat([patientheartrate_, tmp_patientheartrate], axis=1)
patientheartrate
```

```python
#Problem 8
#loại bỏ các cột không cần thiết và giữ lại các thông tin quan trọng trong DataFrame.

columns_to_drop = ['sex_and_time', 'hours_lower', 'hours_upper']
existing_columns = [col for col in columns_to_drop if col in patientheartrate.columns]
patientheartrate = patientheartrate.drop(existing_columns, axis=1)
```

```python
#loại bỏ các hàng có giá trị NaN từ DataFrame patientheartrate
patientheartrate = patientheartrate.dropna()
patientheartrate
```

```python
#11
#cung cấp cho bạn tỉ lệ phần trăm dữ liệu thiếu trên cột 'PulseRate', giúp đánh giá mức độ thiếu sót của dữ liệu trong cột này.
missing_bp = patientheartrate['PulseRate'].isnull().sum() / len(patientheartrate) * 100
```

```python
print(missing_bp)
```

```python
# xử lý các giá trị thiếu trong cột 'PulseRate' bằng cách sử dụng kỹ thuật nội suy để cải thiện chất lượng dữ liệu.
patientheartrate['PulseRate_filled'] = patientheartrate['PulseRate'].fillna(patientheartrate['PulseRate'].interpolate())
```

```python
#đảm bảo rằng tất cả các giá trị trong cột 'PulseRate' đều có thể được biểu diễn như là số
patientheartrate['PulseRate'] = pd.to_numeric(patientheartrate['PulseRate'], errors='coerce')
```

```python
#xử lý dữ liệu thiếu bằng cách sử dụng giá trị trung bình của các giá trị gần nhau
patientheartrate['PulseRate_filled'] = patientheartrate['PulseRate'].fillna(patientheartrate['PulseRate'].rolling(window=2, min_periods=1).mean())
```

```python
#cột 'PulseRate_filled' trong patientheartrate sẽ chứa các giá trị của 'PulseRate' đã được điền các giá trị thiếu theo phương pháp trên.
patientheartrate['PulseRate_filled'] = patientheartrate['PulseRate'].fillna(patientheartrate['PulseRate'].rolling(window=2, min_periods=1).mean().shift(-1))
```

```python
#giá trị 'PulseRate' sẽ được thay thế bằng trung bình của nhóm tương ứng ('Id')
patientheartrate['PulseRate_filled'] = patientheartrate.groupby('Id')['PulseRate'].transform(lambda x: x.fillna(x.mean()))
```

```python
#giá trị 'PulseRate' sẽ được thay thế bằng trung bình của nhóm tương ứng ('Firstname')

patientheartrate['PulseRate_filled'] = patientheartrate.groupby('Firstname')['PulseRate'].transform(lambda x: x.fillna(x.mean()))
```

```python
#đảm bảo rằng mỗi giá trị thiếu trong cột 'PulseRate' sẽ được thay thế bằng giá trị trung bình của toàn bộ cột 'PulseRate'.
patientheartrate['PulseRate_filled'] = patientheartrate['PulseRate'].fillna(patientheartrate['PulseRate'].mean())
```

```python
#đảm bảo rằng mỗi giá trị thiếu trong cột 'PulseRate' sẽ được thay thế bằng giá trị ổn định trong y học  ( 120)
medical_stable_value = 120
patientheartrate['PulseRate_filled'] = patientheartrate['PulseRate'].fillna(medical_stable_value)

```
