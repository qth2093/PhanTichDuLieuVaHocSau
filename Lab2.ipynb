```python
#The thu vien
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
#1
#doc du lieu tu files
dftuyensinh = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
```

```python
#Sắp xếp các dòng của DataFrame 'df' theo giá trị của cột 'DH1'.
dftuyensinh.sort_values(by='DH1')
```

```python
#2
#Sắp xếp các dòng của DataFrame 'df' theo giá trị của cột 'DT' và 'DH1'.
dftuyensinh.sort_values(by=['GT','DH2'])
```

```python
#3
#Tự đặt tên hàm get_mean và scores là tham số đầu vào 
#Hàm sum để tính toán các tham số đầu vào scores
def get_mean(scores):
    return sum(scores)/len(scores)
```

```python
#get the mear score for each subject
result = dftuyensinh['DH1'].sum()

print("\nThe result of applying the function on the dataframe:\n")
print(result)
```

```python
#Sử dụng hàm 'percentile' của NumPy để tính giá trị tại vị trí 25% trong mảng 'g'
def Q1(g):
    return np.percentile(g,25)
```

```python
#Sử dụng hàm 'percentile' của NumPy để tính giá trị tại vị trí 50% trong mảng 'g'
def Q2(g):
    return np.percentile(g,50)
```

```python
#Sử dụng hàm 'percentile' của NumPy để tính giá trị tại vị trí 75% trong mảng 'g'
def Q3(g):
    return np.percentile(g,75)
```

```python
#Sử dụng hàm 'std' của NumPy để tính độ lệch chuẩn của mảng 'g'.
def std(g):
    return np.std(g)
```

```python
#3
#index=['KT']: use cột KT  làm chỉ mục của bảng tổng hợp.

pd.pivot_table(data=dftuyensinh, index='KT', values='DH1', aggfunc=['sum','mean','median','min','max', std, Q1, Q2, Q3])
```

```python
#4
#index=['KT','KV']: Sử dụng hai cột KT, KV làm chỉ mục của bảng tổng hợp.

pd.pivot_table(data=dftuyensinh, index=['KT','KV'], values='DH1', aggfunc=['sum','mean','median','min','max', std, Q1, Q2, Q3])
```

```python
#2
dftuyensinh.loc[(dftuyensinh['T1'] <5), 'phanlopt1'] = 'kem' # values column T1 <5
dftuyensinh.loc[(dftuyensinh['T1'] >=5) & (dftuyensinh['T1'] <7), 'phanlopt1'] = 'trungbinh'# values column T1 5 - <=7
dftuyensinh.loc[(dftuyensinh['T1'] >=7) & (dftuyensinh['T1'] <8), 'phanlopt1'] = 'kha'# values column T1 7 - <=8
dftuyensinh.loc[(dftuyensinh['T1'] >=8), 'phanlopt1'] = 'gioi' # values column T1 >=8
```

```python
#hàm head với 5 hàm đầu tiên
dftuyensinh.head(5)
```

```python
#3
#Nhóm dữ liệu vào các cột ['T1', "phanlopt1"] sử dụng hàm count để tính các giá trị 
#Kết quả đã tính sẽ lưu vào dfphanlop1
dfphanlopt1 = dftuyensinh.groupby(['T1', 'phanlopt1'])[['T1', 'phanlopt1']].count()
```

```python
#Biến đổi các levels của index thành các cột được lưu vào unstackphanlop1
unstackphanlopt1 = dfphanlopt1.unstack()
```

```python
# use unstack va ham head de chay 10 dong dau tien
unstackphanlopt1.head(10)
```

```python
#4
#Chọn cột 'phanlopt1' đã được unstack.
#plot(ylabel='count'): Vẽ biểu đồ dữ liệu từ cột 'phanlopt1' với label trục y là 'count'.
unstackphanlopt1['phanlopt1'].plot(ylabel= 'Count')
```

```python
#5
#['phanlopt1']: Chọn cột 'phanlopt1' từ DataFrame sau khi unstack.
#.plot.bar(): Phương thức plot.bar() của pandas, sử dụng để vẽ biểu đồ cột (bar chart) từ dữ liệu trong cột 'phanlopt1'
unstackphanlopt1['phanlopt1'].plot.bar()
```
